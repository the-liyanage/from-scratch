```python
import numpy as np
```

```python
"""
Step 01: Sigmoid function

- Logistic regression squashes a linear output into a probability [0, 1].
- sigmoid(z) = 1/ (1+e^(-z))
- When z is large --> output ~1 , when z is small --> output ~0

"""

def sigmoid(z):
    return 1/ (1+ np.exp(-z))

"""
Step 02: The Model

- We compute z = X @ weights + bias
- Then:    y_hat = sigmod(z) --> predicted probability

"""

class LogisticRegressionScratch:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.lr = learning_rate
        self.iterations = iterations
        self.weights = None
        self.bias = None

        """
        Step 03: Initialize weights
        - Start weights at 0 
        - bias is also 0 to start 
        """

        def fit(self, X, y):
            n_samples, n_features = X.shape
            self.weights = np.zeros(n_features)
            self.bias = 0

            """
            Step 04: Gradient descent loop
            - We want to minimize the Binary Cross-Entropy loss
            - L = -1/n * Σ[y * log(y_hat) + (1 - y) * (log(1 - y_hat))]

            The gradients to work out to
             dL/dw = 1/n * X^T @(y_hat- y)
             dL/db = 1/n * Σ (y_hat - y)

            """
            for _ in range(self.iterations):
                # forward pass 
                z = X @ self.weights + self.bias  # linear combination
                y_hat = sigmoid(z)                # predicted probabilities 

                # Compute gradients 
                error = y_hat - y
                dw = (1/ n_samples) * X.T @ error
                db = (1/ n_samples) * np.sum(error)

                # Update weights
                self.weights -= self.lr * dw
                self.bias    -= self.lr * db


         """
         Step 05: Predict

        - Get probabilities, then threshold at 0.5:
        - probability >= 0.5 --> class 1
        - probability < 0.5  --> class 0
        """
        def predict_proba(self, X):
            z = X @ self.weights + self.bias
            return sigmoid(z)

        def predict(self, X):
            return (self.predict_proba(X) >= 0.5).astype(int)

"""
Step 06:
 Test it on a simple dataset
"""
       
```
