```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the data 
salary_data = pd.read_csv('data/Salary_Data.csv.xls')

#First 5 rows of the dataset
salary_data.head()
```

```python
print(salary_data)
```

```python
#Checking any missing values
print(salary_data.isnull().sum())
```

```python
plt.scatter(salary_data['YearsExperience'],salary_data['Salary'], color='orange')
plt.xlabel('Years of Experience')
plt.ylabel('Salaries')
```

```python
#Define the features (Independent and Dependent variables)

X = salary_data.iloc[:,0:1]
y = salary_data.iloc[:,-1]
```

```python
#Splitting the dataset into train-set and test-set
#20% goes to testing data-set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state=2)

print(f"Training set size: {X_train.shape}")
print(f"Testing set size: {X_test.shape}")
```

```python
#Training the model 

from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
#train model on the training set

model.fit(X_train, y_train)
```

```python
#Plot all the points
plt.scatter(salary_data['YearsExperience'], salary_data['Salary'])
plt.plot(X_train, model.predict(X_train), color='orange')
plt.xlabel('Years of Experince')
plt.ylabel('Salaries')
```

```python
y_pred = model.predict(X_test)
y_pred
```

```python
y_test.values
```

## Evaluation of Matrics 

### 1. Mean Squared Error(MSE)

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error,r2_score

print("Mean Squared Error: ", mean_squared_error(y_test,y_pred))
```

### 2. Mean Absolute Error(MAE)

```python
print("Mean Absolute Error: ",mean_absolute_error(y_test,y_pred))
```

### 3. Root Mean Squared Error

```python
mse = mean_squared_error(y_test, y_pred)
root_mse = mse**0.5

print("Root Mean Squared:", root_mse)

#Second way of finding the Root Mean Squared
print("Root Mean Squared:", np.sqrt(mse))
```

### 4. R2 Score

```python
print("R2 Score:" , r2_score(y_test, y_pred))
```

### 5. Adjusted R2 Score

```python
r2score = r2_score(y_test, y_pred)
N = X_test.shape[0]

numerator = (1-r2score)*(N-1)
denominator = N-1-1
adjusted_r2score = 1 - (numerator/denominator)
print("Adjusted R2 Scre:" , adjusted_r2score)
```
